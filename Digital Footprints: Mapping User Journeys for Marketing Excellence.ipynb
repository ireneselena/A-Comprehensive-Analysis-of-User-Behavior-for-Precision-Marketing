## Data Distribution and Normality Test

### Histograms of Numerical Data

The histograms below help us understand the data's distribution, highlighting whether the data is skewed or exhibits other non-normal characteristics.


```python
import seaborn as sns
from scipy.stats import shapiro

# Load your DataFrame (assuming it is named 'df')

# List of numerical columns
numerical_columns = [
    'Age', 'Income', 'Total Time Spent', 'Number of Sessions',
     'Video Length', 'Engagement', 'Importance Score',
    'Time Spent On Video', 'Number of Videos Watched', 'Scroll Rate',
     'ProductivityLoss', 'Satisfaction', 'Self Control', 'Addiction Level'
]

# Plot histograms for each numerical column
plt.figure(figsize=(16, 16))
for i, column in enumerate(numerical_columns):
    plt.subplot(5, 4, i + 1)  # Adjust the grid size based on your number of columns
    sns.histplot(df[column], kde=True, color='royalblue')
    plt.title('Histogram of ' + column)
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Perform the Shapiro-Wilk test for normality
normality_results = {}
for column in numerical_columns:
    # Remove NA values if necessary
    data_clean = df[column].dropna()
    if len(data_clean) < 3:
        stat, p_value = float('nan'), float('nan')
        interpretation = "Number of data less than 3, test cannot be performed"
    else:
        stat, p_value = shapiro(data_clean)
        interpretation = "Data is not normally distributed" if p_value < 0.05 else "Data is normally distributed"
    normality_results[column] = {'Statistic': stat, 'p-value': p_value, 'Interpretation': interpretation}

# Convert test results to DataFrame for a neater look
normality_df = pd.DataFrame(normality_results).T
normality_df.index.name = 'Variable'

# Displaying results
print(normality_df)
```

**Normality Test Summary**

- The Shapiro-Wilk test results indicate that all numerical variables in the dataset are not normally distributed. Each variable's p-value is less than the significance threshold of 0.05, confirming a departure from normality. 
- This finding suggests that further analysis should account for non-normal data distributions, possibly through data transformations or the use of statistical methods that do not assume normality. 
- Understanding these distribution characteristics is essential for ensuring accurate and meaningful insights from the data.

### <b><span style='color:darkblue'> Data Creating</span></b>

Now, we are going to create a new column based on `Age` column values to divide them into generations

```python
# Assigning time groups for the trips
def generation(age):
    if 12 <= age < 28:
        return 'Gen Z'
    elif 28 <= age < 44:
        return 'Millenials'
    elif 44 <= age < 60:
        return 'Gen X'
    else:
        return 'Baby Boomers'

# Create a new column for time groups
df['Generation'] = df['Age'].apply(generation)
```

```python
df['Generation'].value_counts()
```

Lastly, we are going to divide the `income` values into several income groups to make it easier to analyze the pattern

```python
# Set the style of the visualization
sns.set(style="whitegrid")

# Create a box plot for the income column
plt.figure(figsize=(8, 2))
sns.boxplot(x='Income', data=df)

# Add title and labels
plt.title('Income Distribution')
plt.ylabel('Income')

# Show plot
plt.show()
```

```python
df['Income'].describe()
```

```python
def income_group(income):
    if income < 38675.25:
        return 'Low Income'
    elif 38675.25 <= income < 58805:
        return 'Lower-Middle Income'
    elif 58805 <= income < 79792.25:
        return 'Middle Income'
    elif 79792.25 <= income < 89676:
        return 'Upper-Middle Income'
    else:
        return 'High Income'

# Apply the function to create a new column for income groups
df['Income Group'] = df['Income'].apply(income_group)

df.head(3)
```
